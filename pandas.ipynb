Apply function

```python
import pandas as pd
def calc_sum(x):
return x.sum()

data = {
  "x": [50, 40, 30],
  "y": [300, 1112, 42]
}

df = pd.DataFrame(data)

x = df.apply(calc_sum)

print(x)
```

Sorting DataFrames

```python
import pandas as pd
data = {
  "score": [50, 40, 30, 40, 20, 60, 70],
  "qualified": [True, False, False, False, False, True, True]
}
df = pd.DataFrame(data)

newdf = df.sort_values(by='score')
```

Removing duplicates

```python
import pandas as pd
unique = df.duplicated((inplace = True))
print(unique)
```
